## 2. Data Quality Assessment & Cleaning


This section evaluates the quality of the dataset by checking for missing values and potential outliers. Appropriate data cleaning decisions are made based on statistical reasoning rather than arbitrary modification.


### 2.1 Missing Values

Missing values can negatively impact statistical analysis and model reliability. Therefore, the dataset is examined for null or missing entries across all features.


```python
# Check for missing values in each column
# This helps determine whether data imputation or removal is necessary

df.isnull().sum()
```

The results show that there are **no missing values** in the dataset.  
However, a median-based imputation strategy is defined for numerical variables as a **preventive and reusable approach**, ensuring robustness if the dataset is extended or updated in future analyses.


```python

numeric_cols = [
    "Financial Loss (in Million $)",
    "Number of Affected Users",
    "Incident Resolution Time (in Hours)"
]

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

### 2.2 Outlier Analysis (No Deletion)

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)).sum()
    print(col, "outliers detected:", outliers)
```

**Decision:**
Outliers are retained because extreme cyber incidents are analytically
significant and represent high-impact real-world events rather than noise.

## 3. Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is conducted to uncover patterns, trends, and insights within the cybersecurity dataset using statistical summaries and visualizations.


### 3.1 Distribution of Attack Types

This section analyzes the frequency distribution of different cyber attack types to identify the most common threats observed in the dataset.


```python
# Plot the distribution of cyber attack types
# This visualization highlights the most frequent attack categories
df["Attack Type"].value_counts().plot(kind="bar", figsize=(8,5))
plt.title("Distribution of Cyber Attack Types")
plt.xlabel("Attack Type")
plt.ylabel("Number of Incidents")
plt.show()
```

### Interpretation

The bar chart illustrates the distribution of different cyber attack types.  
The results indicate that **DDoS and Phishing attacks** are among the most frequently reported incidents, suggesting that they remain dominant threats in the cybersecurity landscape.

Other attack types such as **SQL Injection, Ransomware, and Malware** also appear with significant frequency, highlighting the diversity of cyber threats faced by organizations.

These findings emphasize the need for multi-layered security strategies that address both high-frequency and high-impact attack vectors.


### 3.2 Number of Incidents Per Year

This section examines the annual distribution of reported cybersecurity incidents to identify temporal trends over the period 2015–2024.


```python
# Plot the number of cybersecurity incidents per year
# This visualization highlights temporal trends in incident frequency
df["Year"].value_counts().sort_index().plot(kind="bar", figsize=(8,5))
plt.title("Cybersecurity Incidents Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Incidents")
plt.show()
```

### Interpretation

The bar chart shows a generally **increasing trend** in reported cybersecurity incidents over time, with noticeable growth after 2017.  
This trend may reflect:
- Increased digitalization across industries
- Greater reliance on online services
- Improved reporting and detection mechanisms

Although slight fluctuations are observed in certain years, the overall pattern indicates that cybersecurity threats have become more frequent and persistent in recent years.


### 3.3 Top 10 Most Affected Countries

This analysis identifies the countries with the highest number of reported cybersecurity incidents.


```python
# Plot the top 10 countries with the highest number of incidents
# This helps identify regions most frequently targeted by cyber attacks
df["Country"].value_counts().head(10).plot(kind="bar", figsize=(8,5))
plt.title("Top 10 Most Affected Countries")
plt.xlabel("Country")
plt.ylabel("Number of Incidents")
plt.show()
```

### Interpretation

The chart reveals that countries such as the **UK, Brazil, India, Japan, and France** report the highest number of cybersecurity incidents.

This distribution may be influenced by:
- The size and maturity of digital infrastructure
- High internet penetration and online activity
- More developed cybersecurity reporting frameworks

It is important to note that higher incident counts do not necessarily imply weaker security, but may reflect stronger detection and disclosure practices.


## 4. Univariate Analysis

Univariate analysis is conducted to examine the distribution and statistical characteristics of individual numerical variables within the dataset.

### 4.2 Incident Resolution Time Distribution

```python
# Plot the distribution of incident resolution time
# This histogram shows how long cybersecurity incidents typically take to resolve
plt.hist(df["Incident Resolution Time (in Hours)"], bins=30)
plt.title("Distribution of Incident Resolution Time")
plt.xlabel("Resolution Time (Hours)")
plt.ylabel("Frequency")
plt.show()
```

## 5. Bivariate Analysis

Bivariate analysis is performed to examine relationships between pairs of variables and to identify potential patterns or associations within the dataset.


### 5.1 Financial Loss by Attack Type

```python
# Boxplot of financial loss grouped by attack type
# This visualization compares the distribution of losses across different cyber attacks
df.boxplot(
    column="Financial Loss (in Million $)",
    by="Attack Type",
    rot=45,
    figsize=(8,5)
)

plt.title("Financial Loss by Attack Type")
plt.suptitle("")  # Remove automatic pandas title
plt.xlabel("Attack Type")
plt.ylabel("Financial Loss (Million $)")
plt.show()
```

### Interpretation

The boxplot compares the distribution of financial losses across different cyber attack types.  
While median losses appear relatively similar across attack categories, noticeable variability is observed within each group.

Certain attack types show wider interquartile ranges, indicating higher uncertainty and potential risk.  
This suggests that the financial impact of cyber attacks depends not only on the attack type but also on contextual factors such as target industry and incident response effectiveness.


### 5.2 Financial Loss vs Number of Affected Users

```python
# Scatter plot showing the relationship between affected users and financial loss
# This helps assess whether higher user impact corresponds to greater financial loss
plt.scatter(
    df["Number of Affected Users"],
    df["Financial Loss (in Million $)"],
    alpha=0.5
)

plt.title("Financial Loss vs Number of Affected Users")
plt.xlabel("Number of Affected Users")
plt.ylabel("Financial Loss (Million $)")
plt.show()
```

### Interpretation

The scatter plot reveals a **weak or non-linear relationship** between the number of affected users and financial loss.  
Incidents affecting a large number of users do not always result in proportionally higher financial losses.

This observation indicates that financial impact is influenced by additional factors such as data sensitivity, regulatory penalties, and system criticality, rather than user count alone.


### 5.3 Correlation Analysis

```python
df[[
    "Number of Affected Users",
    "Financial Loss (in Million $)",
    "Incident Resolution Time (in Hours)"
]].corr()
```

### Interpretation

The correlation matrix indicates **very weak linear relationships** between the examined numerical variables.

- Financial loss shows almost no correlation with the number of affected users.
- Incident resolution time has a negligible correlation with financial loss.
- The number of affected users is not strongly associated with resolution time.

These results suggest that cybersecurity impact cannot be explained by a single factor alone and is influenced by multiple contextual and operational variables.


## 6. Trend & Multivariate Analysis

### 6.1 Average Financial Loss Over Time

```python
# Plot the trend of average financial loss over time
# This visualization highlights long-term changes in financial impact
df.groupby("Year")["Financial Loss (in Million $)"].mean().plot(figsize=(8,5))
plt.title("Average Financial Loss Over Time")
plt.xlabel("Year")
plt.ylabel("Average Financial Loss (Million $)")
plt.show()
```

### Interpretation

The line plot illustrates fluctuations in average financial losses over time rather than a strictly increasing or decreasing trend.

These variations may reflect:
- Changes in attack sophistication
- Improvements in security measures
- Differences in reporting standards across years

Overall, the trend suggests that financial impact remains consistently significant over time, emphasizing the persistent economic risk of cybersecurity incidents.


### 6.2 Resolution Time vs Financial Loss

```python
# Scatter plot showing the relationship between resolution time and financial loss
# This helps assess whether longer incidents result in higher financial losses
plt.scatter(
    df["Incident Resolution Time (in Hours)"],
    df["Financial Loss (in Million $)"],
    alpha=0.5
)

plt.title("Financial Loss vs Incident Resolution Time")
plt.xlabel("Resolution Time (Hours)")
plt.ylabel("Financial Loss (Million $)")
plt.show()
```

### Interpretation

The scatter plot shows no clear linear relationship between incident resolution time and financial loss.

Some incidents with short resolution times still result in high financial losses, while others with longer durations cause relatively lower losses.  
This suggests that financial impact depends more on the nature and severity of the attack rather than resolution time alone.
